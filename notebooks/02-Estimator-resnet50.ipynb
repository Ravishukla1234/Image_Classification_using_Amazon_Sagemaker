# Training and Hosting a PyTorch model in Amazon SageMaker

---

## Setup

```python
import sagemaker
import os
from PIL import Image
from matplotlib import pyplot as plt

import numpy as np
sagemaker_session = sagemaker.Session()

bucket = sagemaker_session.default_bucket()
prefix = "pytorch-cnn-cat-dog-example"

role = sagemaker.get_execution_role()

```

---

## Prepare the training data

We have already stored dogs and cats images in the `../input/train` folder

---

## Upload to s3

```python
# from sagemaker.s3 import S3Uploader

# inputs = S3Uploader.upload("../input/train", "s3://{}/{}/data".format(bucket, prefix))
```

```python
inputs = "s3://{}/{}/data".format(bucket, prefix)
inputs
```

---

## Run a SageMaker training job

```python
hyperparameters = {
                    "epochs" : 5 ,
                    "trainable_layer_count" : "6",
                    "lr": 1e-5
                    }
```

```python
from sagemaker.pytorch import PyTorch

estimator = PyTorch(
    entry_point="train.py",
    source_dir = "../src",
    role=role,
    framework_version="1.4.0",
    py_version="py3",
    instance_count=1,
    instance_type="ml.p2.xlarge",
    hyperparameters = hyperparameters
)

estimator.fit(inputs)
```

```python

predictor = estimator.deploy(initial_instance_count=1, instance_type="ml.c5.xlarge")
```

---

# Generate Inference on test set

```python
test_image_directory = "../input/test"

test_image_files = os.listdir(test_image_directory)
```

```python
from PIL import Image
from matplotlib import pyplot as plt

import numpy as np
def plot_img(test_image_files, base_path):
    for img in test_image_files[35:40]:
        image = Image.open(f"{base_path}/{img}")
        image = image.resize((256, 256),Image.BILINEAR).convert("RGB")
        image = np.array(image)
        print(image.shape)
        plt.imshow(image)
        plt.show()        
        outputs = predictor.predict(np.array(image))
        print(outputs)
        if(outputs < 0.5):
            outputs = "cat"
        else:
            outputs = "dog"

        print(outputs)   
```

```python
print(f"Doing infernce on some 5 test images")

plot_img(test_image_files ,test_image_directory )
```
